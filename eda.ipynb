```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df 
```

```python
sns.histplot(df['sepal_length'], bins=30)
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
sns.kdeplot(df['sepal_length'])
plt.title("Density Plot of Sepal Length")
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
sns.violinplot(x='species', y='sepal_length', data=df)
plt.title("Violin Plot of Sepal Length by Species")
```

```python
numeric_df = df.iloc[:, 0:4]
corr_matrix = numeric_df.corr()
corr_matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample stock data (price change %)
data = {
    'TCS': [1.2],
    'INFY': [-0.8],
    'RELIANCE': [2.1],
    'HDFC': [-1.5],
    'ICICI': [0.6]
}

df_stocks = pd.DataFrame(data)
df_stocks
sns.heatmap(df_stocks, annot=True, cmap='RdYlGn', center=0)
plt.title("Stock Market Heatmap (Price Change %)")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
stocks = {
    'TCS': 'TCS.NS',
    'INFY': 'INFY.NS',
    'RELIANCE': 'RELIANCE.NS',
    'HDFC': 'HDFCBANK.NS',
    'ICICI': 'ICICIBANK.NS'
}
data = {}

for name, symbol in stocks.items():
    stock = yf.download(symbol, period="2d")
    stock['pct_change'] = (stock['Close'] - stock['Open']) / stock['Open'] * 100
    data[name] = stock['pct_change'].iloc[-1]

df = pd.DataFrame(data, index=['Yesterday % Change'])
df
sns.heatmap(df, annot=True, cmap='RdYlGn', center=0)
plt.title("Stock Market Heatmap – Yesterday")
plt.show()
import yfinance as yf
import pandas as pd

```
