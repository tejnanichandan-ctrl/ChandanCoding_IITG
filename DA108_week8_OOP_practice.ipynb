# Part 3 – Understanding the `self` Keyword

In Python classes, `self` refers to the **current instance (object)**.

It allows the object to:
- store its own data
- access its own attributes
- call its own methods

Every instance method must have `self` as its first parameter.

## Why Do We Need `self`?

Consider this situation:

We create multiple objects from the same class.

Example:

s1 = Student("Alice",20)  
s2 = Student("Bob",22)

Both objects use the **same class code**, but each object must store **different data**.

`self` helps Python keep track of **which object is being used**.

```python
class Student:

    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello", self.name)


s1 = Student("Alice")
s2 = Student("Bob")

s1.greet()
s2.greet()
```

## What Happens Internally?

When we write:

s1.greet()

Python internally converts it to:

Student.greet(s1)

So the object itself is passed as the first argument.

That object becomes `self` inside the method.

```python
class Student:

    def greet(self):
        print("Object memory address:", self)


s1 = Student()
s2 = Student()

s1.greet()
s2.greet()
```

## Accessing Attributes Using `self`

Attributes must be accessed using `self`.

Correct example:

```python
class Student:

    def __init__(self, name):
        self.name = name

    def show_name(self):
        print(self.name)


s1 = Student("Alice")
s1.show_name()
```

```python
class Student:

    def __init__(self, name):
        self.name = name

    def show_name():
        print(self.name)   # This will cause an error
```

# Part 4 – Common Errors When Creating Classes

When beginners start writing classes, a few mistakes occur frequently.

We will examine the most common ones:

1. Forgetting the `self` parameter
2. Accessing attributes incorrectly
3. Forgetting parentheses when creating objects
4. Not using the `__init__` constructor

## Error 1 – Forgetting the `self` Parameter

Instance methods must always include `self` as the first parameter.

If `self` is missing, Python cannot connect the method to the object.

```python
class Student:

    def __init__(self, name):
        self.name = name

    def print_name():   # Missing self
        print(self.name)


s1 = Student("Alice")
s1.print_name()
```

## Correct Version

```python
class Student:

    def __init__(self, name):
        self.name = name

    def print_name(self):
        print(self.name)


s1 = Student("Alice")
s1.print_name()
```

## Error 2 – Accessing Attributes Outside the Object

Attributes belong to objects.

You must access them using the object name.

```python
class Student:

    def __init__(self, name):
        self.name = name


s1 = Student("Alice")

print(name)   # Incorrect
```

## Correct Way

```python
class Student:

    def __init__(self, name):
        self.name = name


s1 = Student("Alice")

print(s1.name)
```

## Error 3 – Forgetting Parentheses When Creating an Object

When creating an object, we must use parentheses.

Without parentheses, Python assigns the **class itself**, not an instance.

```python
class Student:
    pass


obj = Student   # Missing ()

print(obj)
```

## Correct Way

```python
class Student:
    pass


obj = Student()   # Correct

print(obj)
```

## Error 4 – Not Using the Constructor (`__init__`)

If a class does not define `__init__`, objects can still be created.

But attributes must then be assigned manually.

## Error 4 – Not Using the Constructor (`__init__`)

If a class does not define `__init__`, objects can still be created.

But attributes must then be assigned manually.

```python
class Student:
    pass


s1 = Student()

s1.name = "Alice"
s1.age = 20

print(s1.name)
print(s1.age)
```

## Why Using `__init__` is Better

The constructor ensures that every object starts with proper data.

Better design:

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age


s1 = Student("Alice", 20)

print(s1.name)
print(s1.age)
```

# Part 5 – Creating Your Own Types

Python allows us to create custom objects that behave like built-in types.

Examples of built-in types:
- list
- string
- integer
- dictionary

To make our objects behave similarly, Python provides **special methods** called:

Magic Methods  
or  
Dunder Methods (Double underscore methods)

Example:

__init__  
__str__  
__len__  
__add__

## What Are Magic Methods?

Magic methods are predefined methods surrounded by double underscores.

Example:

__init__   → constructor  
__str__    → string representation of object  
__len__    → defines behavior of len()  
__add__    → defines behavior of +

```python


# Cell 44 (Code)


class Book:

    def __init__(self, title, pages):
        self.title = title
        self.pages = pages


b1 = Book("Python Basics", 300)

print(b1)
```

## Customizing Object Printing with __str__

The `__str__` method controls how an object appears when printed.

```python
class Book:

    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return f"Book: {self.title}, Pages: {self.pages}"


b1 = Book("Python Basics", 300)

print(b1)
```

## Using __len__ to Support len()

The `__len__` magic method defines how the `len()` function behaves for an object.


class Library:

    def __init__(self, books):
        self.books = books

    def __len__(self):
        return len(self.books)


lib = Library(["Python", "Data Science", "AI"])

print(len(lib))

```python
class Library:

    def __init__(self, books):
        self.books = books

    def __len__(self):
        return len(self.books)


lib = Library(["Python", "Data Science", "AI"])

print(len(lib))
```

## Overloading Operators with __add__

Magic methods can also define how operators behave.

Example: `+`

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)


n1 = Number(10)
n2 = Number(5)

result = n1 + n2

print(result.value)
```

## Why Magic Methods Are Powerful

Magic methods allow custom objects to behave like built-in Python objects.

Examples:

len(object)  
print(object)  
object1 + object2  

This enables developers to create **custom data structures** that integrate naturally with Python.

```python
class Movie:

    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def __str__(self):
        return f"{self.title} ({self.rating}/10)"


m1 = Movie("Inception", 9)

print(m1)
```

# Part 7 – Inheritance

Inheritance is a fundamental concept in Object Oriented Programming.

It allows a new class (child class) to inherit attributes and methods from an existing class (parent class).

Benefits:
- Code reuse
- Cleaner program structure
- Easier maintenance

## Parent Class and Child Class

Parent Class (Base Class)
↓
Child Class (Derived Class)

The child class automatically receives the attributes and methods of the parent class.

```python
class Animal:

    def speak(self):
        print("Animal makes a sound")


class Dog(Animal):
    pass


d = Dog()
d.speak()
```

## Adding New Methods in Child Class

A child class can add its own methods while still inheriting parent methods.

```python
class Animal:

    def speak(self):
        print("Animal makes a sound")


class Dog(Animal):

    def bark(self):
        print("Dog barks")


d = Dog()

d.speak()
d.bark()
```

## Using super()

The `super()` function allows the child class to call methods from the parent class.

This is commonly used in constructors.

```python
class Animal:

    def __init__(self, name):
        self.name = name


class Dog(Animal):

    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed


d = Dog("Buddy", "Labrador")

print(d.name)
print(d.breed)
```
