he for Loop — Structural Anatomy
Basic Syntax
for i in range(5):
    print(i)


Let’s break it precisely.

Component 1 — for

Reserved keyword.

Component 2 — i

Iterator variable.

Automatically assigned.

No need to initialize.

Component 3 — range(5)

Generates:

0, 1, 2, 3, 4


Important:

Upper bound is excluded.


```python
for i in range(5):
    print(i)

print(i) # final value of i is 4
```

The range() Function — Full Control
Forms of range:
range(stop)
range(start, stop)
range(start, stop, step)


Examples:

range(5)        → 0,1,2,3,4
range(2,6)      → 2,3,4,5
range(0,10,2)   → 0,2,4,6,8


Important rule:

range(start, stop, step)
→ goes until stop - 1

```python
numbers = [10, 20, 30]

for value in numbers:
    print(value)
```

```python
numbers = [1,2,3,4]

total = 0
for n in numbers:
    total += n

print(total)
```

Common mistake to avoid 

| Mistake                           | Why it happens              |
| --------------------------------- | --------------------------- |
| Expecting range(5) to include 5   | Off-by-one misunderstanding |
| Forgetting indentation            | Syntax error                |
| Not initializing accumulator      | Logic error                 |
| Thinking loop variable disappears | Scope misunderstanding      |


Nested Loops — Loop Inside Loop

Structure :

for i in range(3):
    for j in range(2):
        print(i, j)

Execution Model

Outer loop controls rows.
Inner loop controls columns.

Step-by-step:

When i = 0:

j = 0

j = 1

When i = 1:

j = 0

j = 1

When i = 2:

j = 0

j = 1

Output:

0 0
0 1
1 0
1 1
2 0
2 1

Key Mental Model:

Inner loop completes fully for every outer loop iteration.

This is essential for:

Matrix operations

Pattern printing

Cartesian products

```python
# pattern printing

for i in range(1,6):
    for j in range(i):
        print("*", end=" ")
    print()

for i in range(5,0,-1):
    for j in range(i):
        print("*", end=" ")
    print() 

    
```

3️⃣ break — Immediate Loop Termination

```python
for i in range(10):
    if i == 5:
        break
    print(i)

# about the code - 
# The loop iterates through numbers 0 to 9.
# When i equals 5, the break statement stops the loop.
# Therefore, only numbers 0 to 4 are printed.
```

4️⃣ continue — Skip Current Iteration

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

| break                 | continue            |
| --------------------- | ------------------- |
| Stops loop completely | Skips one iteration |
| Exits loop block      | Stays inside loop   |


5️⃣ while Loop — Condition-Based Iteration

Unlike for, while depends on a condition.

```python
i = 0
while i < 5:
    print(i)
    i += 1
```

Flag-Controlled Loops

Often used in real programs.

```python
flag = True

while flag:
    user = input("Enter number (0 to stop): ")
    if user == "0":
        flag = False
    else:
        print(f"You entered: {user}")   

        
```

| Mistake                         | Why                  |
| ------------------------------- | -------------------- |
| Forgetting `i += 1` in while    | Infinite loop        |
| Using break accidentally        | Loop exits too early |
| Using continue before increment | Skips update logic   |


Strings, Indexing, and Reverse Logic (Without Shortcuts)

This lesson builds directly from loops.

You now learn how iteration works on sequences, especially strings, and how to think in terms of indices rather than shortcuts.

This is foundational for:

Data cleaning

Parsing

Text processing

Algorithmic interviews

```python
s = "CODE"

for i in range(len(s)):
    print(i, s[i])
```

```python
places = ["Assam", "Meghalaya", "West Bengal"]

for p in places:
    print(p[0] + p[len(p)-1])
```

Reverse a String — Without Slicing

```python
s="CODE"
print(s[::-1])
```

```python
s = input("Enter string: ")

for i in range(len(s)-1, -1, -1):
    print(s[i], end="")
```
