```python
print("Hello World")
name = "Ravi"
age = 21

print("Name:", name, "Age:", age)
# Using f-string for formatted output
print(f"Name: {name}, Age: {age}")
# Using string concatenation
print("Name: " + name + ", Age: " + str(age))
# Using the format method
print("Name: {}, Age: {}".format(name, age))
# Using old-style string formatting
print("Name: %s, Age: %d" % (name, age))
```

3️⃣ range() Function
Purpose:

Generates a sequence of numbers.

Very useful in loops.

```python
#range(start, stop, step)
for i in range(5):
    print(i)
for i in range(1, 6):
    print(i)
for i in range(0, 10, 2):
    print(i)
for i in range(10, 0, -1):
    print(i)
# Using range to create a list of numbers
numbers = list(range(1, 11))
print(numbers)
# Using range to iterate over a list
for num in numbers:
    print(num)

# Using range to create a list of even numbers
even_numbers = list(range(0, 21, 2))
```

3️⃣ Lambda Functions (Anonymous Functions)

Now we enter compact function design.

What is a Lambda?

A small, one-line function without a name.

Syntax:

lambda arguments: expression

```python
#Normal function
def normal_function():
    print("This is a normal function.")

def square(x):
    return x * x

square = lambda x: x * x

print(square(5))
```

Why Use Lambda?

Used when:

Function is small

Used temporarily

Passed as argument

Example with sorted():

students = [("Ravi", 85), ("Amit", 92), ("Priya", 78)]

sorted_students = sorted(students, key=lambda x: x[1])

print(sorted_students)


Here:

lambda x: x[1] extracts marks for sorting

Very powerful in data science.

4️⃣ Recursive Functions

Recursion means:

A function calls itself.

Used when problem can be broken into smaller versions of same problem

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(10))
```

5️⃣ Generator Functions

Generators are memory-efficient functions.

They use yield instead of return.

Instead of returning entire list at once,
they generate values one by one.

```python
def get_numbers():
    yield 1
    yield 2
    yield 3
    yield 4

for num in get_numbers():
    print(num)
```

| return                              | yield                   |
| ----------------------------------- | ----------------------- |
| Ends function                       | Pauses function         |
| Returns single value                | Returns sequence lazily |
| Not memory efficient for large data | Memory efficient        |

Real-world analogy:

That mapping concept is exactly what Python dictionaries implement.

```python
student = {
    "name": "Ravi",
    "age": 21,
    "course": "Data Science"
}
print(student["name"])
print(student.get("marks"))
student["marks"] = 95
print(student.get("marks", 0))
student.pop("course")
if "name" in student:
    print("Key exists")
student.keys()
student.values()
student.items()
student.clear()
student.update({"city": "Delhi"})
```
