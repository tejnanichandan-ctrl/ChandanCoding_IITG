What is String Iteration?

Iteration means going through each character one by one.

A string is a sequence of characters, so Python can loop through it.

```python
f = "Python"

for x in f:
    print(x)


for x in f:
    print(x, end="")



for i in range(0, len(f)):
    print(f[i])


for x in [31, 1, 2026]:
    print(x)

for x in ["31", "01", "2026"]:
    print(x)
    for y in x:
        print(y, end=" ")
    print()

```

Lesson 3: String Immutability (VERY IMPORTANT CONCEPT)

This lesson answers one core question that confused you in the lab:

Everything here is exactly from your screenshots + transcript.

1️⃣ What does “Immutable” mean?

Immutable = cannot be changed

In Python:

Strings are immutable

Once created → characters inside cannot be modified

Why Python Made Strings Immutable (Exam + Interview ⭐)

Important reasons:

Security

Passwords stored as strings cannot be modified accidentally

Memory Efficiency

Strings are reused internally (faster, safer)

Predictable Behavior

No side effects in code

👉 This is by design, not a mistake.

```python
s = "Hello World"
print(s.upper())
print(s)
```

```python
#Correct Way to “Change” a String ✔️

#You must create a new string.

s = "Hello World"
s = 'T' + s[1:]
print(s)


s = "Python"
s = s[:2] + 'X' + s[3:]
print(s)

```

```python
#replace() Example (From Your Lab)
s = "Hello World"
s = s.replace("World", "Python")
print(s)
```

```python
l = ['A', 'B', 'C']
l[0] = 'Z'
print(l)
#Lists are Mutable ✔️
#Strings are Immutable ❌
```


String Concatenation (+)
Meaning

Concatenation = joining strings together

```python
a = "Hello"
b = "World"
print(a + b)
print(a + " " + b) # Adding space between
print(a, b) # Using comma adds space by default
print("{} {}".format(a, b)) # Using format method
print(f"{a} {b}") # Using f-strings output - Hello World
print("-".join([a, b])) # Using join method with hyphen output - Hello-World
print("".join([a, b])) # Using join method without space output - HelloWorld
print(a * 3) # Repeating string output - HelloHelloHello
print((a + " ") * 3) # Repeating with space output - Hello Hello Hello
name = "Alice" 
age = 30
print(f"My name is {name} and I am {age} years old.") # Using f-strings output - My name is Alice and I am 30 years old.
print("My name is {} and I am {} years old.".format(name, age)) # Using format method
print("My name is %s and I am %d years old." % (name, age)) # Using old-style formatting
print("My name is " + name + " and I am " + str(age) + " years old.") # Using concatenation output - My name is Alice and I am 30 years old.
text = "The quick brown fox jumps over the lazy dog"
print(text.split()) # Splits at whitespace output - ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
print(text.count("o")) # Counts occurrences of 'o' output - 4
print(text.find("fox")) # Finds index of 'fox' output - 16
print(text.replace("fox", "cat")) # Replaces 'fox' with 'cat' output - The quick brown cat jumps over the lazy dog
print(text.lower()) # Converts to lowercase output - the quick brown fox jumps over the lazy dog
print(text.upper()) # Converts to uppercase output - THE QUICK BROWN FOX JUMPS OVER THE LAZY DOG
print(text.capitalize()) # Capitalizes first letter output - The quick brown fox jumps over the lazy dog
print(text.title()) # Title case output - The Quick Brown Fox Jumps Over The Lazy Dog
print(text.strip()) # Removes leading/trailing whitespace (none here) output - The quick brown fox jumps over the lazy dog
print(text.startswith("The")) # Checks if starts with 'The' output - True
print(text.endswith("dog")) # Checks if ends with 'dog' output - True
multiline = """This is line one. 
This is line two.
This is line three."""
print(multiline) # Prints multi-line string

```

Lesson 6: Looping Through Strings (for loop & range)

This lesson explains how Python reads a string character-by-character.

🔹 Why Looping Through Strings is Important

You use string loops for:

Counting characters

Validating input

Text processing

1️⃣ Looping Directly Over a String (Most Pythonic Way)
Syntax
for variable in string:
    statement

```python
s = "Python"

for ch in s:
    print(ch)
```

```python
s = "Python"

for i in range(len(s)):
    print(s[i])
```

```python
s = "Python"
i = 0

while i < len(s):
    print(s[i])
    i += 1
```

| for loop         | while loop            |
| ---------------- | --------------------- |
| Cleaner          | More control          |
| Preferred        | Manual condition      |
| Less error-prone | Risk of infinite loop |


```python
# common mistake 

while i < len(s):
    print(s[i])   # infinite loop!
   


print(s[len(s)])  # ❌ last index is len(s)-1
```

```python
# Count the number of vowels in a string
s = "Python"
count = 0

for ch in s:
    if ch in "aeiouAEIOU":
        count += 1

print(count)
```

```python
# Reverse string
s = "Python"
rev = ""

for ch in s:
    rev = ch + rev

print(rev)

```

Lesson 1: Why NumPy? (Performance & Vectorization)
🔹 Problem

Python lists are slow for numerical computation.

🔹 Solution

NumPy uses:

C-level execution

Vectorized operations

Contiguous memory

```python
import numpy as np
import time

size = 100000

p_list = list(range(size))
n_array = np.arange(size)

# Python list timing
start = time.time()
p_list = [x + 2 for x in p_list]
print("Python List time:", time.time() - start)

# NumPy array timing
start = time.time()
n_array = n_array + 2
print("NumPy Array time:", time.time() - start)
```

| Python List        | NumPy Array |
| ------------------ | ----------- |
| Loop in Python     | Loop in C   |
| Element-by-element | Vectorized  |
| Slow               | Very fast   |


| `range()`        | `np.arange()`       |
| ---------------- | ------------------- |
| Returns iterator | Returns NumPy array |
| Needs loop       | Vectorized          |
| Slower           | Faster              |


What is Broadcasting? (Core Idea)

Broadcasting means:

NumPy automatically applies operations between arrays of different shapes without using loops.

In simple words:

You take one number

You apply it to every element of an array

NumPy does this internally in C (very fast)

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
result = arr + 10
print(result)

```
