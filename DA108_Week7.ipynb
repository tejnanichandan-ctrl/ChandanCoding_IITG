```python
f = open("sample.txt", "r")

print("First 5 characters:")
print(f.read(5))

print("\nNext 5 characters:")
print(f.read(5))

f.close()
```

```python
f = open("sample.txt", "r")

print(f.read(5))

f.seek(0)

print("\nAfter seek:")
print(f.read(5))

f.close()
```

# Part 3 – Text File Handling

```python
with open("textfile.txt", "w") as f:
    f.write("Line 1: Python\n")
    f.write("Line 2: File Handling\n")
    f.write("Line 3: DA108\n")
```

```python
with open("textfile.txt", "r") as f:
    content = f.read()
    print(content)
```

```python
with open("textfile.txt", "r") as f:
    print(f.readline())
    print(f.readline())
```

```python
with open("textfile.txt", "r") as f:
    lines = f.readlines()

print(lines)
print(type(lines))
```

```python
with open("textfile.txt", "r") as f:
    for line in f:
        print(line.strip())
```

```python
with open("newfile.txt", "w") as f:
    f.write("Hello\n")
    f.write("This is written using write()\n")
```

```python
lines = [
    "Apple\n",
    "Banana\n",
    "Cherry\n"
]

with open("fruits.txt", "w") as f:
    f.writelines(lines)
```

```python
with open("fruits.txt", "a") as f:
    f.write("Mango\n")
```

# Part 4 – Beyond Text Files

```python
import csv

data = [
    ["Name", "Age", "Marks"],
    ["Alice", 21, 88],
    ["Bob", 22, 91]
]

with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
import csv

with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
import json

student = {
    "name": "Alice",
    "age": 21,
    "marks": 88
}

with open("student.json", "w") as f:
    json.dump(student, f, indent=4)
```

```python
import json

with open("student.json", "r") as f:
    data = json.load(f)

print(data)
print(type(data))
```
